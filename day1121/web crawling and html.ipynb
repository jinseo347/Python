# Web Crawling

- Web crawling is the process of indexing data on web pages by using a program or automated script. These automated scripts or programs are known by multiple names, including web crawler, spider, spider bot, and often shortened to crawler.

### 이해할 기술
- HTTP 구성 및 동작
- HTML/CSS/Javascript 구조

## HTTP 
(공유폴더 HTTP 이해 참조)

- 웹을 통해 HTML/CSS/Javascript/text/csv/json/xml 등등의 TEXT 데이터를 교환 목적
- binary data 교환

### Data 형식

- xml: extendable markup language. 굉장히 복잡하고 어려움. a라는 데이터를 스무개의 부가적 정보로 저장. 
    - 정부기관, 보안인증 등에 많이 쓰임. 
        - 
- html: HTML 규정을 따라 만들어짐.  xml의 자손.(축약해놓은 것) a를 표현하기 위해서는 8개의 부가정보가 필요.
- txt: ascii text (unicode text), a는 그냥 a
- csv: 엑셀 사용하면서 ,로 구분. comma seperated value. (',' + text) 양식
    - 그러나 대규모 웹을 개발하다보면 아쉬운 부분이 있음. 구성이 좀 더 필요. 
    - csv는 텍스트만 컴마로 구분 한 것으로 부가적 정보 표기가 어려움. 
        - ex.이름, 나이, 주소(40) : 여기서 40 이해 못함. 
        - a -> a
- json: a 3개 정도의 부가정보가 필요. 웹표준에 있음. 2000년대 이후 많이 씀.
    - https://www.json.org/json-en.html
- binary

- 다양한 데이터 양식을 사용하는 이유.
    - 학생정보를 보낸다고 생각했을 경우 데이터가 어떠한 형식인지 알려주어야 함. 
    - 세부적으로 형식, 크기, 제약사항(?) 등을 데이터와 함께 주고 받을 수 있으면 ㅈ호음. 
    - 텍스트는 일반적으로 문장이나 공백, 단락 등으로 구성되어 있음. 
    - 그러나 일반적으로 데이터라고 했을경우 2차원 데이터. 엑셀의 표 처럼 구성요소가 무엇인지 나타내는 칼럼으로 구성
    - 텍스트로도 가능하지만 구분에 대한 규정등이 필요함으로 다양한 데이터 양식 필요.
